```python
import numpy as np
import pandas as pd
import xml.etree.ElementTree as ET
import matplotlib.pyplot as plt
```

```python
tree = ET.parse("D:/УНИВЕР/ЭВМ (тервер)/padgett.xml")
root = tree.getroot()
tree
```

```python
names = []
for i in tree.findall('MetaNetwork/nodes/nodeclass/node'):
    names.append(i.attrib.get('id'))
names

# source -> target
#PADGM
dfm = pd.DataFrame(np.random.randn(16, 16), index=names, columns=names)
links = tree.findall("MetaNetwork/networks/network[@id = 'PADGM']")[0].findall('link')
for i in links:
    data = i.attrib
    dfm.loc[data.get('source'), data.get('target')] = data.get('value')

#PADGB
dfb = pd.DataFrame(np.random.randn(16, 16), index=names, columns=names)
links = tree.findall("MetaNetwork/networks/network[@id = 'PADGB']")[0].findall('link')
for i in links:
    data = i.attrib
    dfb.loc[data.get('source'), data.get('target')] = data.get('value')
```

```python
def pagerank(M, num_iterations = 100, d = 0.85):
    N = M.shape[1]
    v = np.random.rand(N, 1)
    v = v / np.linalg.norm(v, 1)
    M_hat = (d * M + (1 - d) / N)
    for i in range(num_iterations):
        v = M_hat @ v
    return v
```

```python
padgm = dfm.to_numpy(dtype='float')
div = padgm.sum(axis=1) 
div[-5] += 1
v = pagerank(padgm / div)
prm = v / v.sum()
prm = np.reshape(prm, (1, 16))[0]
```

```python
padgb = dfb.to_numpy(dtype='float')
div = padgb.sum(axis=1) 
div[[0,1,11,12,14]] += 1
v = pagerank(padgb / div)
prb = v / v.sum()
prb = np.reshape(prb, (1, 16))[0]
```

```python
relation = pd.DataFrame({"Marriage" : prm, "Business" : prb, "Total" : (prm + prb) * 0.5}, index=names)
relation
```
