```python
import matplotlib.pyplot as plt
import numpy as np

data = np.loadtxt('http://scipy-lectures.org/_downloads/populations.txt')
#data format: year  hare  lynx  carrot
```

Assume the population of hares in 1920 is unknown. Suggest a way to estimate this value. Compare an estimated value with the true value and print a ratio of the error to the true value.

```python
# probably done

#it seems to me that hare delta and amount of lynxes are looking very similar just inverted
#so let's figure out the arithmetic mean in each year before 1920 of delta hare and lynxes
#and then take the mean of all means. This I will call "zero"
#the prediction will be sum of the hares in 1919 and the "zero"

plt.figure(figsize=(11, 7))
plt.plot(years, np.full(years.size, np.mean((np.hstack((np.array([0]), np.diff(data[:, 1]))) + data[:, 2]) / 2)), 
         label = 'Zero')
plt.plot(years, np.hstack((np.array([0]), np.diff(data[:, 1]))), label = 'Hare delta')
plt.plot(years, data[:, 1], label = 'Hare')
plt.plot(years, data[:, 2], label = 'Lynx')
plt.scatter(1920, data[-1, 1], label = 'truth of hare in 1920')
plt.scatter(1920, np.mean((np.hstack((np.array([0]), np.diff(data[:-1, 1]))) + data[:-1, 2]) / 2) + data[-2, 1], 
            label = 'prediction')
plt.title('Prediction of hares in 1920')
plt.xlabel('Years')
plt.ylabel('Amount')
plt.xticks(years.astype(int) [::2])
plt.legend()
plt.show()

print("truth = {} and prediction = {}".format(data[-1, 1], np.mean((np.hstack((np.array([0]), np.diff(data[:-1, 1]))) + data[:-1, 2]) / 2) + data[-2, 1]))
```

## Task 2 ##
Plot a graph of total explained variance vs number of principal components.

```python
from sklearn.datasets import load_digits
from sklearn.decomposition import PCA

digits = load_digits()

pca_array = [PCA(n_components = i) for i in range(1, 65)]
X_array = [(pca_array[i]).fit_transform(digits.data.astype(float)) for i in range(64)]
pca_array_out = [np.sum(pca_array[i].explained_variance_) for i in range(64)]

plt.figure(figsize = (14, 7))
plt.bar(np.arange(1, 65), pca_array_out)
plt.plot(np.arange(1, 65), np.full(64, pca_array_out[-1] * 0.99), label = '99%')
plt.plot(np.arange(1, 65), np.full(64, pca_array_out[-1] * 0.95), label = '95%')
plt.plot(np.arange(1, 65), np.full(64, pca_array_out[-1] * 0.9), label = '90%')
plt.plot(np.arange(1, 65), np.full(64, pca_array_out[-1] * 0.85), label = '85%')
plt.legend()
plt.title('PCA total variance and number of components')
plt.xlabel('Number of principal components')
plt.ylabel('Total explained variance')
plt.show() 
```

## Task 3 ##
Find a number of principal components for 99%, 95%, 90%, and 85% of explained variance


```python
pc_99 = np.nonzero(pca_array_out > pca_array_out[-1] * 0.99)[0][0]
pc_95 = np.nonzero(pca_array_out > pca_array_out[-1] * 0.95)[0][0]
pc_90 = np.nonzero(pca_array_out > pca_array_out[-1] * 0.90)[0][0]
pc_85 = np.nonzero(pca_array_out > pca_array_out[-1] * 0.85)[0][0]

print("number of principal components for % of explained variance:\n99% = {}\n95% = {}\n90% = {}\n85% = {}".format(pc_99, pc_95, pc_90, pc_85)) 
```

Plot an original image and restored images from 99%, 95%, 90%, and 85% of explained variance.

```python
from sklearn.datasets import load_digits

digits = load_digits()
images = digits.data.reshape(-1, 8, 8)
#i = np.random.randint(0, 1797)
i = 999
print(i)
pca_array = [PCA(n_components = i) for i in (pc_99, pc_95, pc_90, pc_85)]
X_array = [(pca_array[i]).fit_transform(digits.data.astype(float)) for i in range(4)]
X_array_restored = [(pca_array[i].inverse_transform(X_array[i])) for i in range(4)]

fig, ax = plt.subplots(1, 5, figsize = (20, 4))
ax[0].imshow(digits.images[i], cmap='gray')
ax[1].imshow(X_array_restored[0][i].reshape(8, 8), cmap='gray')
ax[2].imshow(X_array_restored[1][i].reshape(8, 8), cmap='gray')
ax[3].imshow(X_array_restored[2][i].reshape(8, 8), cmap='gray')
ax[4].imshow(X_array_restored[3][i].reshape(8, 8), cmap='gray')
ax[0].set_title('Original\nThis is {}'.format(digits.target[i]))
ax[1].set_title('Restored 99%')
ax[2].set_title('Restored 95%')
ax[3].set_title('Restored 90%')
ax[4].set_title('Restored 85%')
plt.show()
```
